```python
import numpy as np
import pandas as pd
```

```python
df =pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv',usecols=['name','fuel','seller_type','transmission','owner'])
df.head()
```

```python
df.drop('name',axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('fuel',axis=1),df['fuel'], test_size=0.2, random_state=42)
```

```python
X_train
```

## Using Scikitlearn

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

## 1 Applying Label Encoder

```python
le = LabelEncoder()
y_train_scaled = le.fit_transform(y_train)
y_test_scaled = le.transform(y_test)
y_train_scaled
```

```python
Oe = OrdinalEncoder()
y_train_traf= Oe.fit_transform(y_train)
y_test_traf = Oe.transform(y_test)
y_train_traf
```
