```python
import pandas as pd
df= pd.read_csv("data.csv")
df
```

```python
df = df[["Pulse",'Calories']]
df
```

## Applying Normalization : Minmax Scalling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler_df = scaler.fit_transform(df)
scaler_df
```

```python
scaler_df = pd.DataFrame(data=scaler_df , columns=df.columns)
scaler_df
```
